```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/mGalarnyk/Tutorial_Data/master/Pandas/CalIt2.data')
```

**RENAME COLUMNS**

```python
df_rename =df= df.rename(columns={'7':'flow_id','07/24/05':'date','00:00:00':'time','0':'count'})
```

```python
df_rename.head()
```

**SELECTING DATA**

```python
df_new=df_rename[(df_rename["date"]=="07/24/05") & (df_rename["flow_id"]==7)]
df_new.head()
```

```python
df_new.sort_values(by=['count'],ascending=False)
df_new.head(10)
```

**APPLY FUNCTION**

```python
df_new.head(10).min(axis=1)
df_new.head(10)
```

```python
df_new.min()
```

```python
df_new.head(10).mean()
```

**GROUPING**

```python
df_top=df[(df["date"]=="07/24/05") & (df["flow_id"]==7)]
df_top.head()
```

```python
df_top.groupby(by='date')['time'].max()
```

```python
df_top.groupby(by='date')['count'].max()
```
